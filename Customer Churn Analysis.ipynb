```python
import matplotlib.pyplot as plt
import pandas as pd

# Count data
count = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to percentage of total (row-wise)
percent = count.div(count.sum(axis=1), axis=0) * 100

# Plot stacked bar
ax = percent.plot(
    kind='bar',
    stacked=True,
    figsize=(5,4)
)

plt.title("Churn by SeniorCitizen (%)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage%")
plt.legend(title="Churn", loc="upper right")

# Add percentage labels
for i, row in enumerate(percent.values):
    cum_height = 0
    for value in row:
        ax.text(
            i,
            cum_height + value/2,
            f"{value:.1f}%",
            ha="center",
            va="center"
        )
        cum_height += value

plt.show()
```

# Customer Churn Analysis – Executive Summary

## Objective

To identify key factors driving customer churn and highlight areas for improving customer retention.

---

## Key Findings

* **Overall churn rate is ~26.5%**, meaning roughly **1 out of every 4 customers leaves the service**.
* **New customers (tenure 1–2 months)** show the highest churn (around **45–50%**), while churn drops sharply as tenure increases.
* Customers on **month-to-month contracts contribute to ~55–60% of total churn**, making contract type a major driver.
* **1-year and 2-year contract customers churn less than 10%**, showing strong long-term retention.
* **Senior citizens show higher churn (~40%+)** compared to non-senior customers (~23–25%).
* Customers **without partners or dependents churn ~10–12% more** than those with family ties.

---

## Service-Based Insights

* **Fiber optic users show the highest churn (~40–42%)**, compared to DSL users (~18–20%).
* Customers **without Online Security churn nearly twice as much** as those who have it.
* **Tech Support has the strongest impact**: customers without tech support account for **~60%+ of churn**.
* Lack of **Online Backup and Device Protection** increases churn by **15–20%**.
* Streaming TV and Movies show **moderate impact**; users of these services churn slightly less but are not primary churn drivers.

---

## Payment & Usage Behavior

* Customers using **Electronic Check show the highest churn (~45%)**.
* Customers using **automatic payment methods (credit card/bank transfer)** show significantly lower churn (**~15–18%**).
* Payment convenience directly influences customer retention.

---

## Business Implications

* Focus retention efforts on **early-stage customers**.
* Encourage customers to shift from **month-to-month to long-term contracts**.
* Promote **bundled value-added services** to increase customer stickiness.
* Provide targeted support for **senior citizens** and high-risk segments.

---

## Conclusion

Customer churn is primarily influenced by **contract type, tenure, and service adoption**. Strategic bundling, long-term plans, and early customer engagement can significantly reduce churn and improve customer lifetime value.

---

## Recommendations & Next Steps

* **Target early-tenure customers (first 1–3 months)** with onboarding support, welcome offers, and proactive engagement.
* **Promote long-term contracts** by offering discounts or incentives to move customers away from month-to-month plans.
* **Bundle value-added services** (Online Security, Tech Support, Backup, Device Protection) to increase perceived value and reduce churn.
* **Focus on high-risk segments** such as senior citizens and fiber optic users with tailored plans and better support.
* **Encourage automatic payment methods** to improve payment convenience and retention.
* Continuously **monitor churn metrics** and use customer feedback to improve service quality.
